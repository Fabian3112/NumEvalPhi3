```python
with open('chaine_of_thought_examples.json','r') as f:
        examples = json.load(f)

template = 'Question : \n {question} Option 1 : {Option1} or Option 2 : {Option2}. lets think step by step.\n \n Answer: {answer} \n \n '

text = ""
for example in examples:
    text += template.format(question=example['question'],Option1=example['Option1'],Option2=example['Option2'],answer=example['answer'])

system_message = 'you are an AI assistant you will decide wether " Option 1 " or " Option 2 " is correct. Think step by step. Here are some examples : \n \n' + text
print(system_message)
```
